```python
# MATPLOTLIB CRASH COURSE
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
x=([1,2,3])
y=([2,4,6])
# fig,ax=plt.subplots(nrows=1,ncols=1)
plt.figure(figsize=(3,3),dpi=150)
plt.title("LINE GRAPH OF X vs Y",color="green",fontdict={"fontname":"Comic Sans MS","fontsize":30})
plt.xlabel("X-AXIS",color="green",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.ylabel("Y-AXIS",color="green",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.plot(x,y,markersize=10,marker="*",linewidth=1,color="green",label="X_Y",linestyle="--",markeredgecolor="red")
plt.legend(loc="upper left",prop={"weight":"bold","size":10,"family":"Comic Sans MS"})
plt.grid(color="green")
plt.show()
```
